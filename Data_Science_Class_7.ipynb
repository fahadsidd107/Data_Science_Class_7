# LOC - Label Based Selection

```python
# data.loc[<row label>,<column label>]
data.loc[:,'title']
```

```python
data.loc[:,['title','country']]
```

```python
data.loc[:,['type']].value_counts()
```

```python
data.loc[1,'cast'].split(', ')
```

```python
len(data.loc[1,'cast'].split(', ')) 
#split ka kaam hai k jahan jahan ', ' wahan wahan array main daal de
```

```python
import numpy as np
data['cast'].isnull().sum()
```

```python
data.replace(to_replace=np.nan,value='None')
```

```python
data.loc[data['title']=='Kota Factory']
```
